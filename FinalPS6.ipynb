# The Interplay of Extreme Heat Events and Air Quality on Heart Disease-Related Hospital Admissions in the United States

Research Question: How does exposure to extreme heat events affect hospital admissions for heart disease in the United States, while controlling for air quality? 

Hypothesis: Exposure to extreme heat events, as measured by high temperature, will increase the risk of hospital admissions for heart disease, with potentially greater effects in certain regions and for certain demographic groups. This relationship will be influenced by air pollution, with higher levels of air pollution exacerbating the health effects of extreme heat events.

Extreme heat event data was sourced from the National Oceanic and Atmospheric Administration (NOAA) National Centers for Environmental Information, downloaded from https://www.ncei.noaa.gov/access/metadata/landing-page/bin/iso?id=gov.noaa.ncdc:C00761.

Cardiovascular disease hospitalization data was sourced from the Centers for Disease Control and Prevention (CDC), downloaded from https://www.cdc.gov/dhdsp/data_statistics/index.htm.

Demographic and socioeconomic data was sourced from the American Community Survey (ACS), downloaded from https://www.census.gov/programs-surveys/acs.

Air quality data was sourced from the Environmental Protection Agency (EPA) Air Quality System (AQS), downloaded from https://www.epa.gov/aqs.

 # Variables
1)	Extreme heat events (Max_temp): Is the average daily maximum air temperature for each county. 

2)	Heart disease hospitalizations (Hospitalizations): Hospitalizations due to heart disease for individuals aged 65 and older (ALL_A_65UP)

3)	Median family income (Income)—Income in the past 12 months (in 2021 inflation-adjusted dollars)

4)	Education level for each state (Education)—Education attainment, percentage of the population 25 years and over with a Bachelor’s degree

5)	Air quality index (Air_quality): The Air Quality Index (AQI) is a numerical scale used to communicate and quantify the levels of air pollution in a specific area. AQI values typically range from 0 to 500, with higher values indicating poorer air quality and greater potential health risks.

    Good (0-50, Green): Air quality is considered satisfactory, and air pollution poses little or no risk.
    Moderate (51-100, Yellow): Air quality is acceptable; however, there may be a moderate health concern for a very small number of people who are unusually sensitive to air pollution.
    Unhealthy for Sensitive Groups (101-150, Orange): Members of sensitive groups (e.g., children, the elderly, and individuals with respiratory or heart diseases) may experience health effects, but the general public is not likely to be affected.
    Unhealthy (151-200, Red): Everyone may begin to experience health effects, and members of sensitive groups may experience more severe health effects.
    Very Unhealthy (201-300, Purple): Health alert: everyone may experience more serious health effects.
    Hazardous (301-500, Maroon): Health warnings of emergency conditions. The entire population is more likely to be affected.

# Model Summary
R-squared: The R-squared value is 0.063, which indicates that my model explains only 6.3% of the variation in hospital admissions. This low value suggests that the model might not be a good fit, and other factors not included in the model could be important in explaining the variation in hospitalizations for heart disease.

F-statistic and Prob (F-statistic): The F-statistic is 1.667, with a p-value of 0.163. Since the p-value is greater than 0.05, there is insufficient evidence to conclude that at least one of the independent variables (Air_quality, Income, Education, Max_temp) has a significant effect on hospitalizations.

Coefficients and p-values for individual variables:

Air_quality: The coefficient for Air_quality is 0.0443, with a p-value of 0.667. This indicates that there is no statistically significant relationship between air quality and hospitalizations after controlling for the other variables in the model.

Income: The coefficient for Income is -0.0002, with a p-value of 0.024. This suggests that there is a statistically significant negative relationship between income and hospitalizations. As income increases, hospitalizations for heart disease tend to decrease, after controlling for the other variables in the model.

Education: The coefficient for Education is 0.1966, with a p-value of 0.192. This indicates that there is no statistically significant relationship between education and hospitalizations after controlling for the other variables in the model.

Max_temp: The coefficient for Max_temp is -0.1358, with a p-value of 0.618. This indicates that there is no statistically significant relationship between maximum temperature (extreme heat events) and hospitalizations after controlling for the other variables in the model.

In summary, my model suggests that income has a significant negative effect on hospitalizations for heart disease, while air quality, education, and maximum temperature do not show significant relationships.
